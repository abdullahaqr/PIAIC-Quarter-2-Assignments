```python
import numpy as np
from PIL import Image
import os
#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
rootDir = 'photos'
arr = np.zeros([20, 200, 200, 3])
def main(): 
    try:
        def num_of_files():
            try:
                for dirName, subdirList, fileList in os.walk(rootDir):
                    fileCount_ = len(fileList)
                print("Total number of files: ",fileCount_)
            except FileNotFoundError:
                print("Sorry file not found.")        
            ################################################################        
        def find_dir():
            try:
                for dirName, subdirList, fileList in os.walk(rootDir):
                    print('Found directory: %s' % dirName)
                    for fname in fileList:
                        print('\t%s' % fname)
            except FileNotFoundError:
                print("Sorry file not found.")
            ############################################
        def img_resize():
            try:
                i=0
                for dirName, subdirList, fileList in os.walk(rootDir):
                    for fname in fileList:
                        filepath = dirName+os.sep+fname
                        if filepath.endswith('.jpg'):
                                img = Image.open(filepath)
                                img_n = img.resize((200,200))
                                arr[i]= img_n
                                i+=1
                print(arr)
            except FileNotFoundError:
                print("Sorry file not found.")
        print(num_of_files())
        print(find_dir())
        print('Numpy Array of resized images :')
        print(img_resize())
        
        ##################################################
    except RuntimeError:
        print("Found an error !")
##########################################################        
if __name__ == '__main__':
    main()
```
